```python
import numpy as np
a= np.arange(10).reshape(5,2)
b= np.arange(10,30,2).reshape(5,2)
c= a-b
print(c)
```

```python
print(c>0)
```

```python
print(b**2)
```

```python
print("Element mult-->\n",a*b)
print("Matrix mult-->\n",np.dot(a,b.T))
```

```python
print(a.sum()) #sum of all elements
print(a.sum(axis=0)) #sum of each column
print(a.cumsum()) #cumulative sum of all elements
print(a.cumsum(axis=0)) #cumulative sum of each column
print(a.min()) #minimum element
print(a.max()) #maximum element
print(a.mean()) #mean of all elements
```
