```python
import pandas as pd
iris=pd.read_csv("D:\Anagha\PAPER\ML_PAPER_PREP\iris.csv")
iris.head()
```

```python
count=iris.groupby('species').species.count()
#or count=iris.groupby('species').size()
count
```

```python
#to first group sepal length column and get the max petal length for each sepal length and then sort it by petal length
best = iris.groupby('sepal_length')['petal_length'].max().sort_index() #sort index is used here since petal length would now be the index and not a col
best #returns a Series
```

```python
#to get the min and max values of sepal length grouped by species
minmax=iris.groupby(['species']).sepal_length.agg([min,max])
minmax
```

```python
#to sort the min and max in desc order
sorted=minmax.sort_values(by=['min','max'], ascending=False)
sorted
```

```python
#to get avg values by species name where index is species and value is the avg
avg=iris.groupby(['species']).sepal_length.mean()
avg
```

```python
desc=avg.describe()
desc
```
