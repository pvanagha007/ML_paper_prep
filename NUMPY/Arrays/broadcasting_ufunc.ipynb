```python
#basic broadcasting example
import numpy as np
a=np.array([1,2,3])
b=2
print(a+b)
```

```python
#outer addition
import numpy as np
a=np.array([1,2,3,4])
b=np.array([10,20,30])
print(a[:,np.newaxis]+b)
print(a[:,np.newaxis]+b[np.newaxis,:])
#print(a[:,np.newaxis]+b[:,np.newaxis])--> will produce error

```

```python
#for vector quantization algorithm--> example
from numpy import array, argmin, sqrt, sum
observation = array([111.0, 188.0])  #weight and height of an athlete
codes = array([[102.0, 203.0],
               [132.0, 193.0],
               [45.0, 155.0],
               [57.0, 173.0]])          # weight and height of 4 clusters
diff = codes - observation    # the broadcast happens here
dist = sqrt(sum(diff**2,axis=-1))
print(diff)
print(dist)                    #shortest distance gives the best matching cluster
print("The best match is index: ",argmin(dist))
```

```python
#universal functions example
#reduce method
import numpy as np
a=np.arange(9).reshape(3,3)
print("Array:\n",a)
print("1.\n",np.add.reduce(a,0))  #column wise addition
print("2.\n",np.add.reduce(a,1))  #row wise addition
print("5.\n",np.add.reduce(a,(0,1)))  #column wise multiplication
print("3.\n",np.add.accumulate(a,1)) #row wise cumulative addition
print("4.\n",np.multiply.accumulate(a,0)) #column wise cumulative multiplication
```

```python
#safe type casting example
mark = {False: ' -', True: ' Y'}
def print_table(ntypes):
    print('X ' + ' '.join(ntypes))
    for row in ntypes:
        print(row, end='')
        for col in ntypes:
            print(mark[np.can_cast(row, col)], end='')
        print()

print_table(np.typecodes['All'])
```
